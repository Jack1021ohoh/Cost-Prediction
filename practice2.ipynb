```python
from skopt import BayesSearchCV
from skopt.space import Real, Integer, Categorical

xgb_params = {
    'seed': seed,
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse',
    'tree_method' : 'gpu_hist',
    'n_jobs' : -1,
    'importance_type' : 'total_gain',
    'grow_policy' : 'lossguide',
    'n_estimators' : 1000
}

spaces = {
   'eta': Real(1e-4, 0.1, prior = 'log-uniform'),
   'min_child_weight': Integer(1, 10),
   'lambda': Real(0, 10),
   'alpha': Real(0, 1),
   'gamma': Real(0, 1),
   'subsample': Real(0.5, 1),
   'colsample_bytree': Real(0.1, 1, prior = 'log-uniform'),
   'max_depth': Integer(5, 20)
}
```

```python
search = BayesSearchCV(XGBRegressor(**xgb_params), search_spaces = spaces, cv = kfold, random_state = seed)
search.fit(x_train.values, y_train.values)

print(search.best_score_)
print(search.best_params_)
```

```python
import optuna
import xgboost as xgb
from sklearn.model_selection import train_test_split

xgb_params = {
    'verbosity': 0,
    'seed': seed,
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse',
    'tree_method' : 'gpu_hist',
    'n_jobs' : -1,
    'importance_type' : 'total_gain',
    'grow_policy' : 'lossguide',
    'n_estimators' : 500
}

x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size = 0.2)
d_train = xgb.DMatrix(x_train, y_train)
d_valid = xgb.DMatrix(x_valid, y_valid)

def objective(trial):

  spaces = {
    'eta': trial.suggest_loguniform('eta', 1e-4, 0.1),
    'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
    'lambda': trial.suggest_uniform('lambda', 0, 10),
    'alpha': trial.suggest_uniform('alpha', 0, 1),
    'gamma': trial.suggest_uniform('gamma', 0, 1),
    'subsample': trial.suggest_uniform('subsample', 0.5, 1),
    'colsample_bytree': trial.suggest_loguniform('colsample_bytree', 0.1, 1),
    'max_depth': trial.suggest_int('max_depth', 5, 15)
  }

  xgb_params.update(spaces)

  sch_prm = xgb.train(xgb_params, d_train, num_boost_round = 100, evals = [(d_valid, 'valid')], verbose_eval = False, early_stopping_rounds = 20)

  return sch_prm.best_score
```

```python
study = optuna.create_study(direction = 'minimize')
study.optimize(objective, n_trials = 500)
```
